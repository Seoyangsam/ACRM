# Question_8: Plot all Belgian stations on a map of Belgium.


```python
# import packages
from geopandas import GeoDataFrame
from pyproj import crs
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
map_df = gpd.read_file('C:/Users/Annelien/Documents/SCHOOL/ACRM/Project NMBS/group7-group-assignment-/Data/BELGIUM_-_Provinces.zip')

map_df.head()
```

```python
stations = pd.read_csv(r'C:\Users\Annelien\Documents\SCHOOL\ACRM\Project NMBS\group7-group-assignment-\Data\stations.csv')

stations =stations[stations['country-code']=='be']

stations.head(5)
```

```python
geo_df = gpd.GeoDataFrame(stations, geometry=gpd.points_from_xy(stations.longitude, stations.latitude), crs='EPSG:4326')
```

```python
fig, ax = plt.subplots(figsize= (10,10))
map_df['geometry'].plot(ax=ax)

geo_df.plot(ax = ax, markersize = 15, marker = 'o', c = 'white', alpha = 1, zorder = 2)
ax.set_title('Belgium stations')
plt.show()
```
