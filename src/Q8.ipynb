# Question_8: Plot all Belgian stations on a map of Belgium.


```python
import src.data_processing.data as data
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
provinces = data.provinces
stations = data.stations
stations =stations[stations['country-code']=='be']
```

```python
geo_df = gpd.GeoDataFrame(stations, geometry=gpd.points_from_xy(stations.longitude, stations.latitude), crs='EPSG:4326')
```

```python
fig, ax = plt.subplots(figsize= (10,10))
provinces['geometry'].plot(ax=ax, color = "#F0FFFF", edgecolor = "darkgrey")
geo_df.plot(ax = ax, markersize = 20, marker = 'o', c = 'red', alpha = 1, zorder = 2)
ax.set_title('Belgian stations')
plt.show()
```
