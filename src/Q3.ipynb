# Question_3: Plot the number of stations per city for cities with multiple stations.


```python
import src.data_processing.data as data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
facilities = data.facilities
```

```python
df = pd.DataFrame(facilities['city'].value_counts())
```

```python
plt.figure(figsize=(10,10))

df[df['city']>1].plot(kind='bar', width= 0.3)
plt.title('Number of stations per city for cities with multiple stations')
plt.xlabel("city")
plt.xticks(fontsize=7.5)
plt.ylabel("number of stations")
plt.legend(loc="upper left")
plt.xticks(rotation= "vertical")
plt.show()
```
