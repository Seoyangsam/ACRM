```python
import src.data_processing.data as data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
new_data3 = pd.merge(data.stations,data.stops,how ='inner',left_on = "URI",right_on = "parent_stop")
new_data3.head(50)
```

```python
#print(new_data3.groupby('name_x')['name_y'].size())
new_data4 = pd.DataFrame(new_data3.groupby('name_x')['name_y'].size())
new_data4 = new_data4.sort_values(by='name_y',ascending=False)
new_data4 = new_data4
plt.figure(figsize=(100,40))
font2 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size' : 200,
}
plt.xlabel("City",font2)
plt.ylabel("Stations",font2)
plt.bar(new_data4.index,new_data4.name_y,color='red')
#ax = plt.gca()
#ax.axes.xaxis.set_ticklabels([])
```

```python
new_data5 = pd.DataFrame(data.travelers)
new_data5 = new_data5.sort_values(by='week',ascending=False)
plt.figure(figsize=(100,50))
new_data5.rename({'Unnamed':'hhh'})
font2 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size' : 200,
}
plt.xlabel("Number of travelers",font2)
plt.ylabel("Stations",font2)
plt.bar(new_data5.Station, new_data5.week, color='red')
```

```python
new_data6 = pd.DataFrame(data.frequency_trips)
new_data6 = new_data6.rename(columns={'Name of the stop':'stop'})
new_data6 = new_data6.sort_values(by='stop',ascending=False)
data.frequency_trips
plt.figure(figsize=(100,40))
font2 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size' : 200,
}
plt.ylabel("Number of trips",font2)
plt.xlabel("Stations",font2)
plt.bar(new_data6.index, new_data6.stop, color='red')
```
